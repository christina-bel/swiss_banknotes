По оси У находится кол-во строчек, у которых Length попала в данный столбец. Но иногда хочется смотреть на долю столбца от всех имеющихся данных. Для этого для каждой строчки нужно задать вес, равный числу, обратному длине рассматриваемого вектора.

```python
# Задаём список из всех (200) элементов, каждый из которых равен 1/200.
w = [1/len(df)]*len(df)
# Передаём этот список функции hist как значение аргумента weights.
df['Length'].hist(weights=w)
```

Данная гистограмма не устраивает тем, что она визуализирует смешанные данные по обоим классам counterfeit и genuine. Хотелось бы видеть на гистограмме разницу в распределении по этим классам.

```python
# По идее, вызов должен быть такой: df.groupby('Status')['Length'].hist(alpha=0.6)
# Но из-за бага https://github.com/pandas-dev/pandas/issues/10756 приходится делать дополнительный вызов plot
df.groupby('Status')['Length'].plot.hist(alpha=0.6)
# Добавляем легенду
plt.legend();
```

Разницы в распределении между фальшивыми и настоящими купюрами практически нет. Но мы построили распределения только одного фактора. Можно построить аналогичным образом гистограммы и для остальных факторов. Если колонок не очень много, можно построить такую картинку, после которой всё станет ясно со всеми распределениями сразу.

```python
from pandas.plotting import scatter_matrix
colors = {'genuine': 'green', 'counterfeit': 'red'}
scatter_matrix(df,
               # размер картинки
               figsize=(6, 6),
               # плотность вместо гистограммы на диагонали
               diagonal='kde',
               # цвета классов
               c=df['Status'].replace(colors),
               # степень прозрачности точек
               alpha=0.2,
              )
```

Видно, что в распределении диагоналей купюр наблюдается очевидное смешение двух нормальных распределений. Теперь, построим гистограммы этого фактора для двух классов купюр.

```python
df.groupby('Status')['Diagonal'].plot.hist(alpha=0.6)
plt.legend(loc='upper left')
```

```python
# Можно построить отдельные диаграммы рассеивания для некоторой пары признаков
df.plot.scatter(x='Top', y='Bottom', c=df['Status'].replace(colors))
```

# Титаник

```python
df=pd.read_csv('http://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
for column in ['Survived', 'Pclass', 'Sex']:
    print(df[column].value_counts(normalize=True))
```

```python
# Представим ту же информацию, но в картинках для отчёта.
# Нужно, чтобы подписи к картинкам отображались корректно по-русски
matplotlib.rc('font', family='Arial')
```

```python
# Строим картинку
df['Survived'].value_counts().plot(kind='bar')
# Добавляем подпись
plt.title(u'Число выживших на Титанике')
plt.xticks([0, 1], ['Not Survived', 'Survived'], rotation=None)
plt.text(-0.1, 275, '61.5%', fontsize=17, color='white')
plt.text(0.9, 150, '38.5%', fontsize=17, color='silver')
```

```python
# Также можно построить круговую диаграмму
fig, ax = plt.subplots(1)
df['Survived'].value_counts().plot(kind='pie', labels=['', ''], ax=ax, autopct='%.2f%%', fontsize=15)
ax.set_title(u'Доля выживших на Титанике')
ax.axis('equal')
ax.legend(labels=[u'Погибли', u'Выжили'], framealpha=1)
ax.set_ylabel('')
plt.tight_layout()
```

```python
matplotlib.style.use('ggplot')
# Гистограммы информативны, но если нужно вывести много гистограмм на одной картинке, то всё смешается. 
# Тогда целесообразно использовать boxplot.

df['Age'].plot(kind='box')
plt.title(u'Распределение возраста')
```
